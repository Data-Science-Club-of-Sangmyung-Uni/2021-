```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
html = urlopen("http://ncov.mohw.go.kr/bdBoardList_Real.do?brdId=1&brdGubun=13&ncvContSeq=&contSeq=&board_id=&gubun=")  

bsObject = BeautifulSoup(html, "html.parser") 


print(bsObject) # 웹 문서 전체가 출력됩니다. 
```

```python
css = bsObject.select_one("#main_maplayout")
#print(css)
cityData = css.find_all(class_="name")
accumulatedData = css.find_all(class_="num")
newData = css.find_all(class_="before")
```

```python
print(cityData)
print(accumulatedData)
print(newData)
```

```python
def deleteBracket(string): #필요없는 문자 제거
  string = string.strip('(')
  string = string.strip(')')
  string = string.strip('+')
  return string
```

```python
import pandas as pd

result = dict()
for city, accumulated, new in zip(cityData, accumulatedData, newData): #모든 데이터 dict에 저장
  result[city.get_text()] = list()
  result[city.get_text()].append(accumulated.get_text())
  temp = deleteBracket(new.get_text()) #필요한 수치 문자열만 추출
  result[city.get_text()].append(temp)

print(result)
  
```

```python
result = pd.DataFrame(result, index=['누적 확진자', '신규 확진자'])
result.to_csv("number of confirmed cases data.csv")
```
