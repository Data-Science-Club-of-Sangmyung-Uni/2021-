```python
import pandas as pd
import numpy as np
```

```python
# e + 00 없애기
pd.options.display.float_format = '{:.1f}'.format
# 모든 행 출력
pd.set_option('display.max_rows', None)
```

```python
# 출처: https://data.cdc.gov/Case-Surveillance/United-States-COVID-19-Cases-and-Deaths-by-State-o/9mfq-cb36
US_covid19 = pd.read_csv("/content/United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv")
US_covid19 = US_covid19.sort_values(by=["submission_date"], axis=0)
US_covid19 = US_covid19.reset_index(drop=True)
US_covid19.head(10)
# tot_cases : 누적 확진자 수
# new_cases : 일일 확진자 수
```

```python
# 출처: https://data.cdc.gov/Vaccinations/COVID-19-Vaccinations-in-the-United-States-Jurisdi/unsk-b7fc
US_vacc = pd.read_csv("/content/COVID-19_Vaccinations_in_the_United_States_Jurisdiction.csv")
US_vacc = US_vacc.sort_values(by=["Date"], axis=0)
US_vacc = US_vacc.reset_index(drop=True)
US_vacc.head(10)
# Administered: 수혜자가 거주하는 관할 구역에 따라 최소 1회 접종을 받은 사람의 총 수
# Series_Complete_Yes : 수혜자가 거주하는 관할 구역을 기준으로 완전 예방 접종을 받은 사람의 총 수(2회 접종 백신 2차 접종 또는 1회 접종 백신 1회 접종)
```

```python
covid19_data = US_covid19.groupby("submission_date")["tot_cases","new_case"].sum()
covid19_data.head(10)
# covid19_data.tail(10)
```

```python
vacc_data = US_vacc.groupby("Date")["Administered","Series_Complete_Yes"].sum()
vacc_data = vacc_data[vacc_data.Administered != 0]
vacc_data.head(10)
# vacc_data.tail(10)
```

```python
# 2020-12-14부터
# 2021-07-31까지
covid19_data = covid19_data["2020-12-14":"2021-07-31"]
vacc_data = vacc_data["2020-12-14":"2021-07-31"]
print(covid19_data.shape, vacc_data.shape)
```

```python
# 두 데이터 합치기
US_covid19_vacc_data = pd.concat([covid19_data, vacc_data],axis=1)
US_covid19_vacc_data.head(30)
# US_covid19_vacc_data.tail(30)
```
